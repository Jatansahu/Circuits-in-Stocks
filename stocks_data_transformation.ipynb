```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import datetime
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv("https://raw.githubusercontent.com/shreya139/Circuits-in-Stocks/data/DataByMinute.csv.xls")
df.columns=["Datetime","Open","High","Low","Close","Volume","Dividends","Stock Splits"]
df.head()
```

```python
scaler = MinMaxScaler()
d=scaler.fit_transform(df[["Open","High","Low","Close","Volume","Dividends","Stock Splits"]])
scaled_df = pd.DataFrame(d)
scaled_df.columns=["Open","High","Low","Close","Volume","Dividends","Stock Splits"]
scaled_df["Datetime"]=df["Datetime"]
cols = list(scaled_df.columns)
scaled_df = scaled_df[[cols[-1]]+cols[:-1]]
scaled_df.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(scaled_df['Datetime'], scaled_df['Open'])
plt.title('Opening Price chart ' + 'Amazon')
plt.show()
```
