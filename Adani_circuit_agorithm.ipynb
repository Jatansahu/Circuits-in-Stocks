```python
from pandas_datareader import data as pdr
import pandas as pd
import yfinance as yf
yf.pdr_override()

from pandas_datareader import data as pdr

import yfinance as yf
yf.pdr_override()


data = pdr.get_data_yahoo("ATGL.NS", start="2019-01-01", end="2023-03-07",interval='1d')
data
```

```python
a=np.zeros((len(data),1))
```

```python
data['high_min_preclose%'] =a 
for i in range(1,len(data)):
  data['high_min_preclose%'][i] = ((data['High'][i] - data['Close'][i-1] )/(data['Close'][i-1]))*100 
data
```

```python
def circuit(column):
    if column["high_min_preclose%"] >=5:
        return 1
    elif column["high_min_preclose%"]<=-5:
        return -1
    else:
      return 0
data = data.assign(circuit=data.apply(circuit, axis=1))
print(data)
```

```python
data.to_csv("circuit.csv")
```
